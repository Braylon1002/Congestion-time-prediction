```python
#################计算出每一天速度之间的欧氏距离##############
def cal_eucl(data):
    distance =[]
    for i in range (18):
        data_1=pd.read_excel('C:\\Users\Admin\\Desktop\\小美赛\\数据\\数据处理\\{}.xlsx'.format(i+1))
        data_sub =data_1 -data
        sub = list(data_sub['速度'].values)
        d =0
        for j in range(len(sub)):
            d =sub[j]**2+d
        distance.append(d)
    return distance 
```

```python
###############最近的数据进行排序
def  sele(data):
    d1 =cal_eucl(data)
    d  = pd.DataFrame(d1)
    d.sort_values(by = 0,ascending =True,inplace = True)
    index = list(d.index)
    index_true =[i+1 for i in index]
    return index_true 
```

```python
###################选取最近的k个数据处理##################
def k_select(k,data):
    index_all = sele(data)
    index = [] 
    for i in range(k):  
        index.append(index_all[i+1])
    return index
```

```python
###########获取time
def time(data):
    data['time'] =0
    for i in list(range(len(data['时间']))):
        t =str(data['时间'][i]).split()[1]
        data.loc[i,'time'] = t
    return data
```

```python
df =time(df)
```

```python
################均方差的计算###########3
def MSE(speed1,speed2):
    speed =speed1-speed2
    return math.sqrt(sum(list(speed**2)))/1080
```

```python
MSE(df['速度'],speed)
```

```python
##################绝对方差#############
def MAE(speed1,speed2):
    speed =speed1-speed2
    return np.mean(list(abs(speed)))

```

```python
MAE(df['速度'],speed)
```
